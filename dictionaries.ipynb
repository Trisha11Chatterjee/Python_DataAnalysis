```python
# Imports
import math
import collections
import numpy as np
import pandas as pd
import matplotlib.pyplot as pp

# Enables inline plotting in Jupyter
%matplotlib inline

# Creating and manipulating dictionaries
capitals = {'United States': 'Washington, DC', 'France': 'Paris', 'Italy': 'Rome'}
print(capitals)

# Accessing dictionary elements
print(capitals['Italy'])

# Adding a new key-value pair
capitals['Spain'] = 'Madrid'
print(capitals)

# Attempting to access a non-existing key
# print(capitals['Germany'])  # Uncomment to see the KeyError

# Checking if keys exist in the dictionary
print('Germany' in capitals, 'Italy' in capitals)

# Merging dictionaries
morecapitals = {'Germany': 'Berlin', 'United Kingdom': 'London'}
capitals.update(morecapitals)
print(capitals)

# Removing a key-value pair
del capitals['United Kingdom']
print(capitals)

# Using tuples as keys in a dictionary
birthdays = {(7, 15): 'Michele', (3, 14): 'Albert'}
print(birthdays[(7, 15)])

# Hashing examples
print(hash('Italy'))
print(hash((7, 15)))

# Empty dictionary and its length
print({})
print(len({}))

# Iterating through dictionary keys
for country in capitals:
    print(country)

# Iterating explicitly using .keys()
for country in capitals.keys():
    print(country)

# Accessing all keys
print(capitals.keys())
print(list(capitals.keys()))

# Iterating through dictionary values
for capital in capitals.values():
    print(capital)

# Iterating through dictionary items
for country, capital in capitals.items():
    print(country, capital)

# Working with sets
continents = {'America', 'Europe', 'Asia', 'Oceania', 'Africa', 'Africa'}  # Duplicates removed automatically
print(continents)

# Checking membership
print('Africa' in continents)

# Adding and removing elements in a set
continents.add('Antarctica')
print(continents)
continents.remove('Antarctica')
print(continents)

# Iterating through a set
for c in continents:
    print(c)
```
