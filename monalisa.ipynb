```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as pp

# Configure inline plotting for Jupyter Notebook
%matplotlib inline

# Define x as a range of values
x = np.linspace(0, 5 * math.pi, 64)

# Calculate sine and cosine values
sinx = np.sin(x)
cosx = np.cos(x)

# Plot sine, cosine, and logarithmic functions
pp.plot(x, sinx, label='sin(x)')
pp.plot(x, cosx, label='cos(x)')
pp.plot(x, np.log(1 + x), label='log(1+x)')
pp.legend()
pp.show()

# Perform calculations and plot more functions
y = sinx * cosx
z = cosx**2 - sinx**2
pp.plot(x, y, label='sin(x) * cos(x)')
pp.plot(x, z, label='cos²(x) - sin²(x)')
pp.legend()
pp.show()

# Perform addition with a slice
w = sinx + 1.5
print("First 10 values of sinx:", sinx[:10])
print("First 10 values of w:", w[:10])

# Plot sinx and w
pp.plot(x, sinx, label='sin(x)')
pp.plot(x, w, label='sin(x) + 1.5')
pp.legend()
pp.show()

# Load and manipulate a grayscale image of the Mona Lisa
monalisa_bw = np.loadtxt('/content/monalisa.txt')  # Ensure the file is in the working directory
print("Shape of Mona Lisa grayscale image:", monalisa_bw.shape)

# Apply gradient along the x-axis
xgrad = np.linspace(0, 1, monalisa_bw.shape[1])
monalisa_xgrad = monalisa_bw * xgrad

# Display original and x-gradient modified images
pp.figure(figsize=(8, 5))
pp.subplot(1, 2, 1)
pp.imshow(monalisa_bw, cmap='gray')
pp.title("Original Mona Lisa")

pp.subplot(1, 2, 2)
pp.imshow(monalisa_xgrad, cmap='gray')
pp.title("Mona Lisa with X Gradient")
pp.show()

# Apply gradient along the y-axis
ygrad = np.linspace(0, 1, monalisa_bw.shape[0])[:, np.newaxis]
monalisa_ygrad = monalisa_bw * ygrad

# Display original and y-gradient modified images
pp.figure(figsize=(8, 5))
pp.subplot(1, 2, 1)
pp.imshow(monalisa_bw, cmap='gray')
pp.title("Original Mona Lisa")

pp.subplot(1, 2, 2)
pp.imshow(monalisa_ygrad, cmap='gray')
pp.title("Mona Lisa with Y Gradient")
pp.show()

# Demonstrating NumPy array operations
a = np.array([0, 1, 2])
b = np.array([-1, -2, -3])
print("Dot product of a and b:", np.dot(a, b))

# Generate a random 3x3 matrix
C = np.random.randn(3, 3)
print("Random 3x3 matrix C:\n", C)

# Matrix-vector multiplication
print("C @ a:", C @ a)
```
