```python
# 02_02: Lists, Tuples, and the Slicing Syntax

import math
import collections

import numpy as np
import pandas as pd
import matplotlib.pyplot as pp

# Enable inline plotting
%matplotlib inline

# List operations
nephews = ["Huey", "Dewey", "Louie"]
print(nephews)
print(len(nephews))

print(nephews[0])
print(nephews[2])
# Accessing out-of-range index will throw an error
# print(nephews[3])

print(nephews[-1], nephews[-2])

for i in range(3):
    nephews[i] = nephews[i] + ' Duck'

print(nephews)

mix_it_up = [1, [2, 3], 'alpha']
print(mix_it_up)

nephews.append('April Duck')
print(nephews)

nephews.extend(['May Duck', 'June Duck'])
print(nephews)

ducks = nephews + ['Donald Duck', 'Daisy Duck']
print(ducks)

ducks.insert(0, 'Scrooge McDuck')
print(ducks)

del ducks[0]
print(ducks)

ducks.remove('Donald Duck')
print(ducks)

ducks.sort()
print(ducks)

reverse_ducks = sorted(ducks, reverse=True)
print(reverse_ducks)

# Slicing lists
squares = [1, 4, 9, 16, 25, 36, 49]
print(squares[0:2])
print(squares[:4])
print(squares[3:])
print(squares[:])
print(squares[0:7:2])
print(squares[-3:-1])

squares[2:4] = ['four', 'nine']
print(squares)

del squares[4:6]
print(squares)

print(len([]))

# Tuples
integers = ('one', 'two', 'three', 'four')
print(integers)
print(integers[-1], integers[1:3])

# Tuples are immutable
# integers[0] = 1

(a, b) = (1, 2)
c, d = 3, 4

# Enumerating lists
for i, duck in enumerate(ducks):
    print(i, duck)

# Function with variable arguments
def three_args(a, b, c):
    print(a, b, c)

my_args = (1, 2, 3)
three_args(*my_args)
```
